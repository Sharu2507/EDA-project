#### CLEAN DATA

```python
# time is in hh.mm.ss format actual format is hh:mm:ss
# to convert time to the format using
# regex false gives a instruction to pandas to treat only dots and replace it
df['Time']=df['Time'].str.replace('.',':',regex= False)
df.head()
```

```python
# date and time 
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], dayfirst=True) # dayfirst= pandas work in mm/dd/yyyy format but the datset has dd/mm/yyyy
df.head(2)
```

```python
df['Day']=df['DateTime'].dt.day_name()
df.head(2)
```

#### HANDLING MISSING VALUES

```python
df.isnull().sum()
```

```python
df.isna().sum()
```

```python
# pollution values or any values should not be negative and that value is treated as missing value
df['CO(GT)'].value_counts()
```

```python
print(df['CO(GT)'].unique())
```

```python
df.replace(-200,pd.NA,inplace=True)
df.isna().sum().sort_values(ascending=False)
```

| Method       | Use Case                                         | Example                                     |
| ------------ | ------------------------------------------------ | ------------------------------------------- |
| `linear`     | Time series with smooth trend (default and safe) | âœ…                                           |
| `time`       | When index is a datetime and intervals vary      | Best for irregular time data                |
| `polynomial` | When data curves more than straight line         | `interpolate(method='polynomial', order=2)` |
| `spline`     | Smooth curves, use for gentle variations         | `interpolate(method='spline', order=2)`     |
| `nearest`    | Fill using nearest known value                   | For categorical-like sequences              |
| `ffill`      | Fill from the previous value                     | When trends stay flat for a while           |
| `bfill`      | Fill from the next value                         | Similar to ffill                            |


#### first identify the use case
1. If the line is smooth and gently rising/falling, use linear.
2. If the data curves, try spline or polynomial.
3. If values stay flat for a while, use ffill.

```python
# error will come if we plot directly since we are having non numeric value NAN
# so concverting non nyumeric junk to real nan values
# with coerce invalid entries becomes nan
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
df['CO(GT)'].plot(figsize=(12,4), title='CO Levels Over Time')
plt.show()
```

```python
df['NO2(GT)'].plot(figsize=(12,4),title="NO2 Levels over Time",color="Green")
plt.show()
```

```python
df['NOx(GT)'].plot(figsize=(12,4),color="yellow",title="NO Levels over time")
plt.show()
```

```python
df['C6H6(GT)'].plot(figsize=(12,4),color="pink",title="ozone Levels over time")
plt.show()
```

```python
df['CO(GT)'] = df['CO(GT)'].interpolate() # used interpolate to remove nan values 
df['CO(GT)'].isna().sum()
```

```python
df['CO(GT)'].plot(figsize=(12,4), title='CO Levels Over Time')
plt.show()
```
