```python
from spotipy.oauth2 import SpotifyClientCredentials
import spotipy
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
# Set up Client Credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id='b8b5512ac05949ba95eac62339c99f7a',  # Replace Client ID
    client_secret='cbf1633ce1af4d81ababdc4a88e2c135'  # Replace Client Secret
))
```

```python
# Full track URL (example:Mr.Brightside by killers)
track_url = "https://open.spotify.com/track/003vvx7Niy0yvhvHt4a68B"
```

| Function       | What it does                                   |
| -------------- | ---------------------------------------------- |
| `re.search()`  | Finds the first match in the string            |
| `re.match()`   | Checks if the pattern matches at the **start** |
| `re.findall()` | Returns a list of all matches                  |
| `re.sub()`     | Replaces text matching the pattern             |
| `re.split()`   | Splits a string using the pattern              |


```python
# Extract track ID directly from URL using regex
track_id = re.search(r'track/([a-zA-Z0-9]+)', track_url).group(1)
track_id
```

```python
# Fetch track details
track = sp.track(track_id)
print(track)
```

```python
# Extract metadata
track_data = {
    'Track Name': track['name'],
    'Artist': track['artists'][0]['name'],
    'Album': track['album']['name'],
    'Popularity': track['popularity'],
    'Duration (minutes)': track['duration_ms'] / 60000 # the duration is given in milli seconds hence dividing by 60000 to obtain in minutes
}
```

```python
# Display metadata
print(f"\nTrack Name: {track_data['Track Name']}")
print(f"Artist: {track_data['Artist']}")
print(f"Album: {track_data['Album']}")
print(f"Popularity: {track_data['Popularity']}")
print(f"Duration: {track_data['Duration (minutes)']:.2f} minutes")
```

```python
# Convert metadata to DataFrame
df = pd.DataFrame([track_data])
print("\nTrack Data as DataFrame:")
print(df)
```

```python
# Save metadata to CSV
df.to_csv('spotify_track_data.csv', index=False)
```

```python
# Visualize track data
features = ['Popularity', 'Duration (minutes)']
values = [track_data['Popularity'], track_data['Duration (minutes)']]
```

```python
plt.figure(figsize=(8, 5))
plt.bar(features, values, color='skyblue', edgecolor='black')
plt.title(f"Track Metadata for '{track_data['Track Name']}'")
plt.ylabel('Value')
plt.show()
```

```python
plt.plot(features,values,color='green')
plt.title(f"Track Metadata for '{track_data['Track Name']}'")
plt.ylabel("'Value")
plt.grid(True)
plt.show()
```

```python
plt.pie(values,labels=features,autopct='%1.1f%%')
plt.title(f"Track Metadata for '{track_data['Track Name']}'",fontdict={'fontsize':21 ,'fontname':'Times new roman','fontweight':'bold','color':'blue'})
plt.savefig('piechart.jpg',dpi=300) # clarity or image quality (min 300)
plt.figure(figsize=(8,5))                      
plt.show()
```
