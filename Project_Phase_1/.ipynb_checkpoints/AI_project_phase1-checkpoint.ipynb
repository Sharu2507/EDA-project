```python
# creating a group of salary ranges and job title
s_r = df.groupby(['job_title', 'experience_level'])['salary_usd'].agg(['min', 'max'])

# adding  range column
s_r['range'] = s_r['min'].astype(str)+ '-'+ s_r['max'].astype(str)

# Reset index to make the column titles in the correct index
s_r = s_r.reset_index()
```

#### CREATING PLOTS AND GRAPHS

##### LINE PLOT

```python
# creating plot for minimum salary ranges for 1st 5 jobs based on exp == entry level
# assigning plots
x = jobs[:5].tolist()
y = min_sr[:5].tolist() # converting to list for easy plotting and adding proper labels

# altering the figure size
plt.figure(figsize =(13,5))

# plotting the graph
plt.plot(x,y,color ="#0ceb1b" , linestyle = '--' , marker = "o" , markerfacecolor = "navy" , markeredgecolor = "navy" , label = "minimum salary")

# assigning labels
plt.xlabel("Job Title", fontdict = {"fontname":"Comic Sans MS" ,"fontsize":14 , "fontweight":"bold"}) 
plt.ylabel("Salary Range",fontdict = {"fontname":"Comic Sans MS" ,"fontsize":14 , "fontweight":"bold"})

# xticks values alteration
plt.xticks(rotation = 45,fontname = "Comic Sans MS" ,fontsize=8 , fontweight = "bold")
plt.yticks(fontname = "Comic Sans MS" ,fontsize=8 , fontweight = "bold")

# assiging grid values
plt.grid(True , linestyle = "--" , color = "black" , alpha = 0.6)

# adding legend
plt.legend(loc ="upper right", frameon = True , borderpad = 0.9 )

# adding the label value above marker
for i in range(len(x)):
    plt.text(x[i] , y[i]+50 , str(y[i]), ha = "center" , fontsize = 8 , fontweight = "bold",
             bbox={'facecolor':'white', 'edgecolor':'black', 'boxstyle':'round,pad=0.3'})
# adding title
plt.title("Minimum Salary Range ",fontdict = {"fontname":"Comic Sans MS" ,"fontsize":16 , "fontweight":"bold"})

plt.show()
```

##### MULTIPLE LINE PLOT

```python
# gng to plot a graph for 1st 10 job titles with all experience and minimum salary range
# creating plots
x = jobs[:10].tolist() # converting to list for easy access
en = min_sr[:10].tolist()
ex = min_ex[:10].tolist()
mi = min_mi[:10].tolist()
se = min_se[:10].tolist()

# assigning figure size
plt.figure(figsize =(16,7))

# plotting
plt.plot(x,en,color="#0ceb1b",marker ='o',markerfacecolor ="#0ceb1b",markeredgecolor ="#0ceb1b", label = "Entry level" )
plt.plot(x,ex ,color="#02f2da",marker ='o',markerfacecolor ="#02f2da",markeredgecolor ="#02f2da", label = "Executive level" )
plt.plot(x,mi, color="#f7eb0a",marker ='o',markerfacecolor ="#f7eb0a",markeredgecolor ="#f7eb0a", label = "Mid level")
plt.plot(x,se,color="#d4b6d9",marker ='o',markerfacecolor ="#d4b6d9",markeredgecolor ="#d4b6d9", label = "Senior level")

# altering x ticks
plt.xticks(rotation = 35,fontname = "Comic Sans MS" ,fontsize=12)
plt.yticks(fontname = "Comic Sans MS" ,fontsize=12)

# adding x and y label
plt.xlabel("Job Title" , fontname = "Comic Sans MS" ,fontsize=18 , fontweight = "bold")
plt.ylabel("Salary Range",fontname = "Comic Sans MS" ,fontsize=18 , fontweight = "bold")

# adding labels above the marker 
for i in range(len(x)):
    plt.text(x[i], en[i] + 1700, str(en[i]), ha='center', fontsize=10,
             bbox=dict(facecolor='white', edgecolor='green', boxstyle='round,pad=0.4'))

    plt.text(x[i], ex[i] + 1700, str(ex[i]), ha='center', fontsize=10,
             bbox=dict(facecolor='white', edgecolor='cyan', boxstyle='round,pad=0.4'))

    plt.text(x[i], mi[i] + 1700, str(mi[i]), ha='center', fontsize=10,
             bbox=dict(facecolor='white', edgecolor='orange', boxstyle='round,pad=0.4'))

    plt.text(x[i], se[i] + 1700, str(se[i]), ha='center', fontsize=10,
             bbox=dict(facecolor='white', edgecolor='purple', boxstyle='round,pad=0.4'))

# adding legend i want the legend to be placed in the bottom left outside the frame in a flat manner as there is no space in upperright
plt.legend( 
    bbox_to_anchor=(-0.1, -0.35),  # X=0 (left), Y=-0.25 (below plot)
    loc='lower left',           # anchor top-left of the legend box
    facecolor='white',
    edgecolor='black',
    ncol=2,  # to spread in the four columns
    fontsize=12,
)

# adding title
plt.title("Minimum Salary in USD and Jobs",fontdict = {"fontname":"Comic Sans MS" ,"fontsize":22 , "fontweight":"bold"})

plt.show()
```

##### BAR CHART

```python
# comparing minimum and maximum salary of entry level position for the 1st 5 job titles

# creating plots
x = jobs[:5].tolist() # converting to list for easy access
en = min_sr[:5].tolist()
en_max = max_sr[:5].tolist()
job = jobs[:5]

# assigning figure size
plt.figure(figsize =(12,7))
w = 0.3 # width

# i am gng to create a side by side bar so need positions 
bar1 = np.arange(len(x)) # this gives a list of [0,1,2,...]
bar2 = [i+w for i in bar1] # the concept is : we have to add w for each position of bar to get the bar side ways

# plotting
plt.bar(bar1, en , w , label = "min salary", color = 'lightsalmon',hatch = 'o',edgecolor = 'honeydew')
plt.bar(bar2 , en_max , w , label ="max salary", color ='lightcoral',hatch = '/',edgecolor = 'honeydew')

# setting x ticks and y ticks
 
# altering x ticks
plt.xticks(ticks=bar1,labels=job, rotation = 35,fontname = "Comic Sans MS" ,fontsize=10) # because label is not printing as it is assigned to a anumeric
plt.yticks(fontname = "Comic Sans MS" ,fontsize=10)

# setting title
plt.title("Min and Max Salary",fontname = "Comic Sans MS" ,fontsize=20 , fontweight = "bold")

#x and y label
plt.xlabel("Job Title",fontname = "Comic Sans MS" ,fontsize=12 , fontweight = "bold")
plt.ylabel("Salary range (USD)",fontname = "Comic Sans MS" ,fontsize=12 , fontweight = "bold")

#setting labels
for i in range(len(x)):
    plt.text(bar1[i], en[i] + 3000, str(en[i]), ha='center', fontsize=9,
             bbox=dict(facecolor='white', edgecolor='orange', boxstyle='round,pad=0.4'))
    plt.text(bar2[i], en_max[i] + 3000, str(en[i]), ha='center', fontsize=9,
             bbox=dict(facecolor='white', edgecolor='pink', boxstyle='round,pad=0.4'))
# adding legend
plt.legend(bbox_to_anchor=(1, 1.15),loc = "upper right", borderpad = 0.9 , frameon= True,edgecolor = "black")

# adding grid
plt.grid(axis = "y" ,linestyle ="--",color ="black",alpha =0.6)

plt.show()
```

##### PIE CHART

```python
# Creating pie chart for maximum salary percent for the 1st 5 job title

# creating plots
x = jobs[:5].tolist() # converting to list for easy access
en_max = max_sr[:5].tolist()


# figure
plt.figure(figsize=(8,8)) 

#colors
color = ['pink','#bc8ad1','#0bd9d5','#8cf70a','gold']

# explode
explode = [.01,.01,0.01,0.01,0.1]

# plotting
plt.pie(en_max, labels = x , colors = color , autopct ="%.1f%%" , explode = explode , shadow = True)

# title
plt.title("Maximum Salary in USD for Jobs",fontname = "Comic Sans MS" ,fontsize=20 , fontweight = "bold")

#legend
plt.legend(bbox_to_anchor = (-0.5,1),ncol = 2 ,loc = "upper left",frameon = True , borderpad = 0.9 , fontsize = 9 )

plt.show()
```
